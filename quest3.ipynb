#### 1.  <b>corona.xlsx 파일엔 1월20일 부터 1월30일까지의 중국과 타국에서의 COVID19 확진자 추이가 담겨있다.</b>

```python
import pandas as pd #데이터 프레임 이용을 위한 pandas를 임포트하고 pd로 사용 

val_kor=pd.read_excel('D:/dataframe/corona.xlsx', usecols="A, B, C", index_col=0)# corona.xlsx파일을 읽어오되 0번 컬럼을 인덱스로 지정하여 불러옴  
val_kor["한국"]=1,1,2,3,4,7 #한국이란 column을 추가하고 값 대입
val_kor.to_csv("D:/dataframe/covid19.csv", encoding="utf-8-sig")#추가한 컬럼을 반영하여 covid19.csv파일로 저장(인코딩 없을 시 한글 깨짐)
kor_csv=pd.read_csv("D:/dataframe/covid19.csv", index_col=0)#covid19.csv 파일을 읽어오되 0번 컬럼을 인덱스로 지정하여 불러옴
world=pd.read_excel('D:/dataframe/covid19_new.xlsx', index_col=0, usecols="A,B,C,D")#world에 원본 내용을 저장(usecols를 통해 사용할 컬럼을 지정)
china=pd.read_excel('D:/dataframe/covid19_new.xlsx', index_col=0, usecols="A,B")#china에 중국 내용만 저장(usecols를 통해 사용할 컬럼을 지정)
tha=pd.read_excel('D:/dataframe/covid19_new.xlsx', index_col=0, usecols="A,C")#tha에 타국 내용만 저장(usecols를 통해 사용할 컬럼을 지정)
korea=pd.read_excel('D:/dataframe/covid19_new.xlsx', index_col=0, usecols="A,D")#korea에 한국 내용만 저장 (usecols를 통해 사용할 컬럼을 지정)

newco=pd.ExcelWriter('D:/dataframe/covid19_new.xlsx', engine='openpyxl')#ExcelWriter를 사용하여(시트별 내용을 저장하기 위해) covid19_new.xlsx이름으로 저장 
 

world.to_excel(newco, sheet_name='sheet1')#sheet1에 world내용 저장
china.to_excel(newco, sheet_name='sheet2')#sheet2에 china내용 저장
tha.to_excel(newco, sheet_name='sheet3')#sheet3에 tha내용 저장
korea.to_excel(newco, sheet_name='sheet4')#sheet4에 korea내용 저장
 
newco.save()#작성 완료 후 파일 저장
```

#### 2. <b>auto-mpg.csv 파일은 자동차의 연비에 대한 데이터 파일이다.</b>
    - 다음에 맞추어서 해당 데이터셋에 columns name을 설정하고, auto-mpg-2.csv로 저장하기
 
    mpg(연비)
    cylinders(실린더수)
    displacement(배기량)
    horsepower(출력)
    weight(차중)
    acceleration(가속능력)
    modelYear(출시년도)
    originNumber(제조국번호)
    name(모델명)
 
    - 80년대에 생산된 모델의 모델명만을 출력하라
    - mpg(연비)와 horsepower(출력) 열의 표준편차값을 구하라
    - horsepower(출력) 열의 표준편차는 구할 수 없다. 이유를 서술하라.
    - modelYear(출시년도)를 최신순에 따라 데이터를 정렬해서 auto-mpg-3.csv로 저장하라

```python
import pandas as pd #데이터 프레임 이용을 위한 pandas를 임포트하고 pd로 사용

auto = pd.read_csv('D:/dataframe/auto-mpg.csv') # auto-mpg.csv 파일을 읽어옴
column_name = ['mpg(연비)', 'cylinders(실린더수)', 'displacement(배기량)', 'horsepower(출력)', 'weight(차중)', 'acceleration(가속능력)', 'modelYear(출시년도)', 'originNumber(제조국번호)', 'name(모델명)']#추가할 컬럼 리스트에 저장 
data = auto.values.tolist() #values.tolist()를 통해 DataFrame을 리스트형태로 변환
new_auto = pd.DataFrame(data, columns=column_name )#new_auto columns 인자에 생성한 리스트 변수들을 넣어줌으로써 컬럼명 생성
new_auto.to_csv('D:/dataframe/auto-mpg-2.csv', mode = 'w',index=False, encoding="utf-8-sig") #해당 내용을 auto-mpg-2.csv로 저장하되 index를 False를 줘서 자동으로 인덱스가 추가되지 않게한다.
matrix=pd.DataFrame(new_auto)
provi = matrix['modelYear(출시년도)'] >= 80 #ModelYear(출시년도) 컬럼에 값들이 80보다 크거나 같은 값만 출력(80년대 출시된 모델만)

print(matrix[provi]['name(모델명)']) #슬라이싱을 통해서 provi를 만족하는 MODEL의 이름만 출력
mpg_std=matrix['mpg(연비)'].std()# std()를 통해서 mpg(연비) 컬럼 열의 표준편차를 구함 
#horse_std=matrix['horsepower(출력)'].std() # horsepower(출력) 컬럼 값 중 공란이 ?로 저장되어 있어서 value에러가 발생하여 표준편차를 구할 수 없음

print('\n mpg(연비)의 표준편차 :' , mpg_std) #mpg 표준편차 출력
# print(horse_std) value에러로 출력 x
power=matrix.sort_values('modelYear(출시년도)', ascending=False)# 최신순은 내림차순이므로 ascending을 False로 하여 내림차순으로 정렬
power.to_csv('D:/dataframe/auto-mpg-3.csv', mode = 'w',index=False, encoding="utf-8-sig")#정렬된 내용을 auto-mpg-3.csv형태로 저장
 
```

#### 3. <b>남북한발전전력량.xlsx 파일엔 남한과 북한의 발전전력별 연간통계량이 저장되어있다.</b>
    - 해당 데이터를 읽어와서 남한과 북한의 수력 및 화력발전과 그 합계에 대한 연간 통계량만을 분리해서 남북한수력및화력발전전력량.csv로 저장하라. 

```python
import pandas as pd # 데이터 프레임 이용을 위한 pandas를 임포트하고 pd로 사용

korea=pd.read_excel('D:/dataframe/남북한발전전력량.xlsx') # 남북한발전전력량.xlsx파일을 읽어옴
korea_val=pd.DataFrame(korea)
korea_copy=korea.copy()# copy를 통해 korea_copy에 복사본 저장
korea_1 = korea_copy.drop(index=[3,4,8],axis=0) # 사용하지 않을 남한원자력,남한신재생,북한원자력에 해당하는 행들을 지움, axis=0은 행방향으로 동작 만약 axis=1일 경우는 열방향

re_korea=korea_1.T # 필요없는게 제거된 dataframe을 전치
re_korea[0]=re_korea[1]+re_korea[2] # 단순 제거이므로 합계 수정을 위해 전치행렬에 1,2 컬럼, 즉 남한 수력,화력의 합계를 더하여 합계에 해당하는 0 컬럼에 저장
re_korea[0][0]='남한' # 더하는 과정 nan값의 의해 달라진 부분을 수정
re_korea[0][1]='합계' # 수력화력으로 바뀐 값을 합계로 수정
re_re_korea=re_korea.T # 수정 작업이 완료되었으므로 다시 전치시켜 원래 형태로 돌려줌

re_re_korea.to_csv('D:/dataframe/남북한수력및화력발전전력량.csv', mode = 'w',index=False, encoding="utf-8-sig") # 수정된 파일을 남북한수력및화력발전전력량.csv파일로 저장하되 index를 False로 자동으로 추가되지 않게함.
```

#### 4. <b>다음을 만족하는 20년 달력 데이터 프레임을 만드시오.</b>
        original    Year    Month    Day    weekday    weekend    season    
    0   20-01-01     20       1       1       Wed         0       winter
    1   20-01-02     20       1       2       Thr         0       winter
    2   20-01-03     20       1       3       Fri         0       winter
    3   20-01-03     20       1       4       Sat         1       winter
    
    ...   ...        ...     ...     ...      ...        ...        ...   
    
    365 20-12-31     20       12      31      Thr         0       winter
    

```python
import pandas as pd # 데이터 프레임 이용을 위한 pandas를 임포트하고 pd로 사용
x=pd.date_range('2020-01-01',periods=366, freq='D') #x에 20년1월1일부터 하루 간격으로 1년치의 날짜생성

cale = pd.DataFrame({'original':x, # cale이란 데이터 프레임을 생성 후 original 컬럼에는 1년치의 날짜저장
                   'Year':x.year, # year를 통해 년도만 분리하여 Year컬럼에 년도만 저장
                   'Month':x.month, # Month를 통해 월만 분리하여 Month컬럼에 월만 저장
                    'Day':x.day, # day를 이용하여 일만 분리하여 Day컬럼에 일자만 저장
                    'Weekday':x.weekday, # weekday를 이용하여 월요일은 0부터 일요일은 6에 해당하는 값들을 weekday컬럼에 저장
                    'weekend':'', # weekend 컬럼 생성
                    'season':'' # season 컬럼 생성
                    })

cale['Weekday'] = cale['Weekday'].map({0:'Mon', #cale['weekday'](DataFrame).map을 통해 데이터프레임에서 입력데이터를 원하는 내용으로 변환 0은 월요일
                             1:'Tue', # 1은 화요일
                             2:'Wed', # 2는 수요일
                             3:'Thr', # 3은 목요일
                             4:'Fri', # 4는 금요일
                             5:'Sat', # 5는 토요일
                             6:'Sun'}, # 6은 일요일
                            )

cale.loc[cale['Weekday'] == ('Sat') , 'weekend'] = '1'  # loc를 통해 weekday 값이 Sat(토요일)인 경우 해당 행에서 weekend 컬럼에 해당하는 값에는 1을 대입
cale.loc[cale['Weekday'] == ('Sun') , 'weekend'] = '1' # loc를 통해 weekday 값이 Sun(일요일)인 경우 해당 행에서 weekend 컬럼에 해당하는 값에는 1을 대입
cale.loc[cale['Weekday'] != ('Sun' or 'Sat') , 'weekend'] = '0' # 그 외에 경우 즉 평일인 경우는 weekend 컬럼에 0을 대입

cale.loc[cale['Month'] == (12) , 'season'] = 'winter' # loc를 통해 weekday값이 12인 경우 해당 행에서 season 컬럼에 해당하는 값에는 winter(겨울)을 대입
cale.loc[cale['Month'] == (1) , 'season'] = 'winter' # loc를 통해 weekday값이 1인 경우 해당 행에서 season 컬럼에 해당하는 값에는 winter(겨울)을 대입
cale.loc[cale['Month'] == (2) , 'season'] = 'winter' # loc를 통해 weekday값이 2인 경우 해당 행에서 season 컬럼에 해당하는 값에는 winter(겨울)을 대입
cale.loc[cale['Month'] == (3) , 'season'] = 'spring' # loc를 통해 weekday값이 3인 경우 해당 행에서 season 컬럼에 해당하는 값에는 spring(봄)을 대입
cale.loc[cale['Month'] == (4) , 'season'] = 'spring' # loc를 통해 weekday값이 4인 경우 해당 행에서 season 컬럼에 해당하는 값에는 spring(봄)을 대입
cale.loc[cale['Month'] == (5) , 'season'] = 'spring' # loc를 통해 weekday값이 5인 경우 해당 행에서 season 컬럼에 해당하는 값에는 spring(봄)을 대입
cale.loc[cale['Month'] == (6) , 'season'] = 'summer' # loc를 통해 weekday값이 6인 경우 해당 행에서 season 컬럼에 해당하는 값에는 summer(여름)을 대입
cale.loc[cale['Month'] == (7) , 'season'] = 'summer' # loc를 통해 weekday값이 7인 경우 해당 행에서 season 컬럼에 해당하는 값에는 summer(여름)을 대입
cale.loc[cale['Month'] == (8) , 'season'] = 'summer' # loc를 통해 weekday값이 8인 경우 해당 행에서 season 컬럼에 해당하는 값에는 summer(여름)을 대입
cale.loc[cale['Month'] == (9) , 'season'] = 'fall' # loc를 통해 weekday값이 9인 경우 해당 행에서 season 컬럼에 해당하는 값에는 fall(가을)을 대입
cale.loc[cale['Month'] == (10) , 'season'] = 'fall' # loc를 통해 weekday값이 10인 경우 해당 행에서 season 컬럼에 해당하는 값에는 fall(가을)을 대입
cale.loc[cale['Month'] == (11) , 'season'] = 'fall' # loc를 통해 weekday값이 11인 경우 해당 행에서 season 컬럼에 해당하는 값에는 fall(가을)을 대입


cale # 완성된 20년 달력 DataFrame출력
```
