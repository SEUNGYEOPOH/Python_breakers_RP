```python
import pandas as pd
import numpy as np
```

```python
dfm = pd.read_csv('auto-mpg.csv', header=None)
dfm.columns=['mpg(연비)' ,'cylinders(실린더 수)', 'displacement(배기량)', 'horsepower(출력)', 'weight(차중)', 'acceleration(가속력)', 
             'model_year(출시년도)', 'origin_number(제조국 번호)', 'name(모델 명)']
dfm
```

```python
###삭제 
dfm.replace('?',np.nan,inplace=True)
dfm.dropna(inplace=True)
dfm.info()
```

```python
dfm.replace('?',np.nan,inplace=True) #? nan값 대체
dfm = dfm.astype({'horsepower(출력)' : np.float}) #float으로 변환

mean_power = dfm['horsepower(출력)'].mean()# 평균계산
mean_power#결측치제외통계

dfm['horsepower(출력)'].fillna(mean_power, inplace=True)

dfm.describe()
```
